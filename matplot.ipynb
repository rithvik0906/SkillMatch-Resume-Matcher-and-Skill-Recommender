```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x 
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)
plt.show()
```

```python
student={"Bob":87,"Matt":56,"Sam":27}
```

```python
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color='g')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,marker='*',c='g',s=100)
plt.scatter(x,b,marker='.',c='r',s=200)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='g',s=100)
plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',c='r',s=200)
plt.show()
```
