```python
import pandas as pd
import re
```

```python
resume_df = pd.read_csv("datasets/resume.csv")
jobdesc_df = pd.read_csv("datasets/job_description.csv")
```

```python
resume_clean = resume_df[['ID', 'Resume_str', 'Category']].copy()
jobdesc_clean = jobdesc_df[['job_title', 'job_description', 'sector']].copy()
```

```python
resume_clean.dropna(subset=['Resume_str'], inplace=True)
jobdesc_clean.dropna(subset=['job_description'], inplace=True)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text.strip()
```

```python
resume_clean['clean_text'] = resume_clean['Resume_str'].apply(clean_text)
jobdesc_clean['clean_text'] = jobdesc_clean['job_description'].apply(clean_text)
```

```python
resume_clean.to_csv("clean_resume.csv", index=False)
jobdesc_clean.to_csv("clean_job_description.csv", index=False)
```
