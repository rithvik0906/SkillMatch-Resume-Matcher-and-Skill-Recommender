```python
import pandas as pd
import spacy
import re
```

```python
resume_df = pd.read_csv("clean_resume.csv")
job_df = pd.read_csv("clean_job_description.csv")
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
skill_keywords = [
    'python', 'java', 'c++', 'sql', 'machine learning', 'deep learning', 
    'data analysis', 'data visualization', 'excel', 'tableau', 'power bi',
    'communication', 'teamwork', 'leadership', 'cloud', 'aws', 'azure',
    'react', 'javascript', 'node', 'html', 'css', 'nlp', 'statistics'
]
```

```python
def extract_skills(text):
    text = str(text).lower()
    found_skills = [skill for skill in skill_keywords if re.search(r'\b' + re.escape(skill) + r'\b', text)]
    return list(set(found_skills))
```

```python
resume_df['resume_skills'] = resume_df['clean_text'].apply(extract_skills)
job_df['job_skills'] = job_df['clean_text'].apply(extract_skills)
```

```python
print(resume_df[['Category', 'resume_skills']].head(5))
print(job_df[['job_title', 'job_skills']].head(5))
```

```python
resume_df.to_csv("nlp_resume_skills.csv", index=False)
job_df.to_csv("nlp_job_skills.csv", index=False)
```
